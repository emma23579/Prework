```python
# Q2.1

x = [12,43,4,1,6,343,10, 34, 12, 93, 783, 330, 896, 1, 55]

for item in x:
    if item % 3 == 0:
        print(item)

stringlist = []

for item in x:
    item1 = str(item)
    stringlist.append(item1)
print(stringlist)

for string in stringlist:
    if string[-1] == '3':
        print(string)
        
print(min(x))
print(max(x))
```

```python
# Q3.1

word_freq = {'love': 25, 'conversation': 1, 'every': 6, "we're": 1, 'plate': 1, 'sour': 1, 'jukebox': 1, 'now': 11, 'taxi': 1, 'fast': 1, 'bag': 1, 'man': 1, 'push': 3, 'baby': 14, 'going': 1, 'you': 16, "don't": 2, 'one': 1, 'mind': 2, 'backseat': 1, 'friends': 1, 'then': 3, 'know': 2}

for value in word_freq:
    if word_freq[value] < 3:
        print(value)
        
print("\n")

all_values = word_freq.values()

for value in word_freq:
    if word_freq[value] == max(all_values):
        print(value)
      
```
